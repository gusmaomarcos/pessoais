```python
import pandas as pd
import matplotlib.pyplot as plt

# Define your custom x-axis range and tick labels here
x_min = 150
x_max = 4500
x_ticks = [150, 1000, 2000, 3000, 4500]  # Example tick labels, customize as needed

# Load data from the Excel file
#file_path = 'LDPDs_FTIR.xlsx'  # Update with your file path
sheet_name = 'dados'  # Update with your sheet name
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract wavenumbers (assuming they are in the first column)
x_values = df.iloc[:, 0]

# Create a multi-curve stacked lines by y offsets graph
plt.figure(figsize=(10, 6))

# Define a y-offset for each curve
y_offsets = [0, 0.1, 0.2, 0.3, 0.4]

# Iterate through absorbance columns (excluding the wavenumber column)
for i, column in enumerate(df.columns[1:], start=1):
    y_data = df[column]
    #+ y_offsets[i-1]
    plt.plot(x_values, y_data, label=f'Sample {i}')

# Set the x-axis range and custom x-tick labels
plt.xlim(x_min, x_max)
plt.xticks(x_ticks)

# Add labels, title, and legend (optional)
plt.xlabel('Wavenumber (cm^-1)')
plt.ylabel('Absorbance')
plt.title('Multi-Curve Stacked Lines by Y Offsets')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
